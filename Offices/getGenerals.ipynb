```python
import re
import pandas as pd
```

```python
# 拜 etc. 為? office
reg_appoint    = re.compile(r"([拜出除舉命辟起請入署選召徵版補]為?|[兼領行加遷徙轉歷]|進[號位]|追贈)(\w+?)(將軍)")
```

```python
fin = 'songshu.all.zht.txt'
generals = []
cnt = 0
with open(fin, "r", encoding="utf-8") as fi:
    for line in fi:
        cnt += 1
        matches = reg_appoint.findall(line)
        if len(matches) > 0:
            for match in matches:
                print(f"line {cnt}: {match}")
                generals.append([cnt] + list(match))
fi.close()
```

```python
df_generals = pd.DataFrame(generals, columns=['line','appoint-verb','general','jiangjun'])
df_generals.head()
```

```python
df_generals.to_excel("generals.xlsx")
```
