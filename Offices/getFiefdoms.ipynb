```python
import re
import pandas as pd
```

```python
reg_feng    = re.compile(r"封([^為，。：、「」]+?)([王公侯伯子男])(?!主)")         # 封(fiefdom)(title)
reg_feng_wei = re.compile(r"封(\w+?)為([^為，。：、「」]+?)([王公侯伯子男])(?!主)") # 封(person)為(fiefdom)(title)
```

```python
fin = 'songshu.all.zht.txt'
fiefdoms = []
cnt = 0
with open(fin, "r", encoding="utf-8") as fi:
    for line in fi:
        cnt += 1
        # regex 1
        matches1 = reg_feng.findall(line)
        if len(matches1) > 0:
            for match in matches1:
                print(f"line {cnt}: {match}")
                fiefdoms.append([cnt, None] + list(match))
        # regex 1
        matches2 = reg_feng_wei.findall(line)
        if len(matches2) > 0:
            for match in matches2:
                print(f"line {cnt}: {match}")
                fiefdoms.append([cnt] + list(match))
fi.close()
```

```python
df_fiefdoms = pd.DataFrame(fiefdoms, columns=['line','person','fief','title'])
df_fiefdoms.head()
```

```python
df_fiefdoms.to_excel("fiefdoms.xlsx")
```
